```python
import pandas as pd
import requests
import json

def swap(x):
    """ Convert a name from format in the contributors table to the format in the url
    """
    x = x.lower()
    lst = x.replace(".", ",").split(',')
    parsed = ""
    if len(lst) == 2:
        parsed = lst[1][1:] + "-" + lst[0]     
    else:
        if lst[2] == ' dr':
            parsed = lst[2][1:] + "-" + lst[1][1:] + "-" + lst[0]
            #print(parsed)
        elif lst[2] in [' von', ' von der']:
            if lst[1] == ' dr':
                lst = lst[:2]
                print(lst)
                parsed = ""
            else:
                if len(lst) == 3:
                    parsed = lst[1][1:] + "-" + lst[2][1:] + "-" + lst[0][1:]
                    print(parsed)
                else:
                    print(lst)
                    parsed = ""
        else:
            if lst[1] == ' dr':
                if len(lst) == 3:
                    parsed = lst[1][1:] + "-" + lst[2][1:] + "-" + lst[0]
                    #print(parsed)
                elif (lst[2] == ' med') and (len(lst) == 4):
                    parsed = lst[1][1:] + "-" + lst[2][1:] + "-" + lst[3][1:] + "-" + lst[0]
                    #print(parsed)
                else:
                    print(lst)
                    parsed = ""
    return parsed.replace(" ", "-")
```

```python
# read and parse the donations table
df = pd.read_csv("/Users/grisha.oryol/Downloads/result.csv")
df['Betrag'] = pd.Series([x[:-2].replace(",", ".") for x in df['Betrag']]).astype(float)
df["parsed"] = [swap(x) for x in df['Geldgeber']]
```

```python
def get_flattened(dt):
    """ extract the relevant information from a politician's profile at abgeordnetenwatch.de
    """
    plain_keys = ['degree', 'first_name', 'last_name', 'gender', 'birthyear', 'education', 'profession']
    flattened = pd.Series({key: dt['profile']['personal'][key] for key in plain_keys})
    flattened['city'] = dt['profile']['personal']['location']['city']
    flattened['country'] = dt['profile']['personal']['location']['country']
    flattened['postal_code'] = dt['profile']['personal']['location']['postal_code']
    flattened['state'] = dt['profile']['personal']['location']['state']
    flattened['party'] = dt['profile']['party']
    flattened['parliament'] = dt['profile']['parliament']['name']
    
    if len(dt['profile']['constituency']):
        flattened['constituency'] = dt['profile']['constituency']['name']
        flattened['constituency_won'] = dt['profile']['constituency']['won']
    else:
        flattened['constituency'] = "no_constituency"
        flattened['constituency_won'] = "no_constituency"
        
    if len(dt['profile']['list']):
        flattened['list'] = dt['profile']['list']['name']
        flattened['list_won'] = dt['profile']['list']['won']
    else:
        flattened['list'] = "no_list"
        flattened['list_won'] = "no_list"
    
    return flattened



who_donated = {}
for name in df["parsed"]:
    try:
        if len(name):
            r = requests.get("https://www.abgeordnetenwatch.de/api/parliament/bundestag/profile/{}/profile.json".format(name))
            if r.status_code != 404:
                print(name)
                dt = json.loads(r.text)
                who_donated[name] = get_flattened(dt)
    except KerboardInterrupt:
        break
    except:
        print("something went wrong for " + name)
df_abg = pd.DataFrame(who_donated).T
```

```python
# merge the information from abgeordnetenwatch from information from the contributors table
merged = pd.merge(df_abg, df, left_index = True, right_on = "parsed")
```
